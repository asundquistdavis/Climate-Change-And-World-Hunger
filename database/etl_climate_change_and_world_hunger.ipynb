```python
global_temperature_file = "../Resources/GlobalTemperatures.csv"
global_temperature_df = pd.read_csv(global_temperature_file)
global_temperature_df.head()
```

```python
global_temperature_df.dropna()
```

```python
s = global_temperature_df["dt"] = pd.to_datetime(global_temperature_df["dt"])
global_temperature_df["year"] = global_temperature_df["dt"].dt.year
global_temperature_df
```

```python
global_temperature_final_df  = global_temperature_df.groupby("year", as_index = False).mean()
global_temperature_final_df
```

```python
global_temperature_final_df = global_temperature_final_df[(global_temperature_final_df["year"] >= 1992) & (global_temperature_final_df["year"] <= 2013)]
global_temperature_omega_df = global_temperature_final_df[["year", "LandAverageTemperature", "LandAverageTemperatureUncertainty"]]
global_temperature_omega_df = global_temperature_omega_df.rename(columns = {"LandAverageTemperature": "temperature", "LandAverageTemperatureUncertainty": "uncertainty"})
global_temperature_omega_df
```
